```python
import pandas as pd
import matplotlib.pyplot as plt
# Membuat persamaan dan membaca pada file csv
curah_hujan = pd.read_csv('curah_hujan1.csv')

# Menampilkan data csv
curah_hujan
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Membuat figure dengan w=11 dan h=6
plt.figure(figsize=(11,6))

# Membuat jarak berurutan antar chart
x = np.arange(1,13)

# Membuat data untuk multicart
tahun = curah_hujan['Bulan']
tahun_1981_1990 = curah_hujan['1981-1990']
tahun_1991_2000 = curah_hujan['1991-2000']
tahun_2001_2010 = curah_hujan['2001-2010']

# Membuat bar chart
plt.bar(x, curah_hujan['1981-1990'], width=0.25, label='1981-1990', color='pink',)
plt.bar(x + 0.25, curah_hujan['1991-2000'], width = 0.25, label='1991-2000', color='red',)
plt.bar(x + 2 * 0.25, curah_hujan['2001-2010'], width = 0.25, label='2001-2010', color='orange',)

# Membuat garis horizontal pada sumbu y dengan kelipatan 20
for i in range(20, 700, 20):
  plt.axhline (y=i, color='grey', linestyle='-', linewidth=0.3, zorder=1)

# Membuat garis horizontal pada sumbu y dengan kelipatan 100
for i in range(100, 700, 100):
  plt.axhline (y=i, color='black', linestyle='-', linewidth=0.2, zorder=1)

# Membuat sumbu y maksimal 700
plt.ylim(0,700)

# Membuat label pada sumbu x dan y
plt.xlabel('Bulan', fontsize=11, fontweight='bold')
plt.ylabel('Curah Hujan (mm)', fontsize=11, fontweight='bold')

# Membuat legend (kotak informasi)
plt.legend(loc='upper center', ncols=3, frameon = False, bbox_to_anchor=(0.5,1.1))

# Membuat titik bar char di tengah
plt.xticks(x + 0.25, x)

# Menampilkan hasil
plt.show()
```

```python
suhu = pd.read_csv('suhu_permukaan.csv')
suhu
```

```python
# Membuat Persamaan dari data csv
waktu = suhu['Waktu']
asimilasi = suhu['Asimilasi']
observasi = suhu['Observasi']
non_asimilasi = suhu['Non Asimilasi']

# Membuat figure dengan w=12 & h=6
plt.figure(figsize = (12,6))

# Membuat multiply plot
plt.plot(waktu, asimilasi, color='blue', marker='o', linewidth=2.5, label='Asimilasi')
plt.plot(waktu, observasi, color='orange', marker='o', linewidth=2.5, label='Observasi')
plt.plot(waktu, non_asimilasi, color='red', marker='o', linewidth=2.5, label='Non Asimilasi')

# Membuat garis horizontal pada sumbu y
for celcius in range(20,27):
  plt.axhline(y=celcius, color='grey',linestyle='-', linewidth=0.4)

# Memberikan judul
plt.title('Suhu Permukaan', color= 'k', fontsize=13)

# Memberikan label untuk sumbu x dan y
plt.xlabel('Waktu', color='k', fontsize=11,)
plt.ylabel('Celcius (C)', color='k', fontsize=11)

# Memberikan pada sumbu x per titik rotasi = 15
plt.xticks(rotation=15)

# Memberikan limit min dan max pada sumbu y
plt.ylim(20,27)

# Membuat legend
plt.legend(loc='center right', frameon=False, bbox_to_anchor=(1.2, 0.5))

# Menampilkan hasil
plt.show()
```
